```python

comidas = []
bebidas = []
alumnos = []


# agregar alumnos en fila
def alumno(n):



    if n>0:
        alum=input("Ingresa el nombre del alumno:  ")
        alumnos.append(alum)
        n -= 1
        alumno(n)
    else:
        print("Todos los datos se han guardado   ")

#agregar comidas en fila
def comida(n):


    if n > 0:
        com=input("Nombre de la comida preparada:  ")
        comidas.append(com)
        n -= 1
        comida(n)
    else:
        print("Todos los datos se han guardado")

#Agregar bebisdas en fila
def bebida(n):

    if n > 0:
        beb=input("Nombre de la bebida preparada:  ")

        bebidas.append(beb)
        n -= 1
        bebida(n)
    else:
        print("Todos los datos se han guardado")

#hacer une busqueda por alumno
def buscar_alumno():
    alum=input("Ingrese el nombre del alumno:   ")

    print(f" {alum} está {alumnos.count(alum)} veces en la lista")

#hacer una busqueda por comida
def buscar_comida():
    com=input("Ingrese el nombre de la comida:  ")

    print(f" {com} está {comidas.count(com)} veces en la lista")


#hacer busqueda por bebida
def buscar_bebida():
    beb=input("Ingrese el nombre de la bebida:  ")

    print(f" {beb} está {bebidas.count(beb)} veces en la lista")


#servir comida a los alumnos
def asignar():

    for alumno,comida,bebida in zip(alumnos,comidas,bebidas):

      print(" {0} :almorzo :{1}  y tomo  :{2}".format(alumno,comida,bebida))


#imprimir listas  por orden o normal
def imprimir():
    print("1. Orden original")
    print("2. Orden alfabético")
    r =int(input(":"))
    if r == 1:
        print("Alumnos:", alumnos)
        print("Comidas:", comidas)
        print("Bebidas:", bebidas)
    elif r == 2:
        print("Alumnos:", sorted(alumnos))
        print("Comidas:", sorted(comidas))
        print("Bebidas:", sorted(bebidas))
    else:
        print("Opción inválida")


#Agregar alumno al inicio
def inicio_alumno():
    alum=input("Nombre del alumno")

    comidas.insert(0,alum)

#Metodo para agergar comida al inicio
def inio_comida():
    com=input("Nombre de la comida preparada")

    comidas.insert(0,com)
#MMetodo para agregar bebida al inicio
def inicio_bebida():
    beb=input("Nombre de la bebida preparada")

    bebidas.insert(0,beb)


# metodo para eliminar las listas
def borar_listas():
  for i in range (len(alumnos)):
          alumnos.pop()

  for i in range (len(comidas)):
          comidas.pop()

  for i in range (len(bebidas)):
          bebidas.pop()

  print("\nlistas borradas con exito ")



def menu():
    op = 0
    while op != 12:
        print("\n")
        print("1. Ingresar alumno")
        print("2. ingresar comida ")
        print("3. ingresar bebidas")
        print("4. agregar alumno al inicio")
        print("5. agregar comida al inicio")
        print("6. agregar bebida al inicio")
        print("7. Buscar si existe un nombre de alumno en la fila")
        print("8. Buscar si existe comida en la fila y cuántas veces existe")
        print("9. Buscar si existe bebida en la fila y cuántas veces existe")
        print("10. Asignar comida y bebida a alumnos")
        print("11. imprimir listas")
        print("12.borar listas ")
        print("13. Salir ")

        op = int(input(" eleige opcion :"))
        if op == 1:
            n=int(input("Número de alumnos a ingresar:  "))

            alumno(n)
        elif op == 2:
            n=int(input("Número de comidas a preparar:  "))

            comida(n)
        elif op == 3:
            n=int(input("úmero de bebidas a preparar:  "))

            bebida(n)
        elif op == 4:
            inicio_alumno()
        elif op == 5:
            inio_comida()
        elif op == 6:
            inicio_bebida()
        elif op == 7:
            buscar_alumno()
        elif op == 8:
            buscar_comida()
        elif op == 9:
            buscar_bebida()
        elif op == 10:
            asignar()
        elif op ==11:
            imprimir()
        elif op ==12:
           borar_listas()

        else:
           break


menu()
```
